```python
Image = cv2.imread("C:\\Users\\MARWA\\Desktop\\sem 5\\4-B_M\\Tasks\\original.png")
image_or=cv2.cvtColor(Image, cv2.COLOR_BGR2RGB)
plt.imshow(image_or)
plt.show()
```

```python
rows, cols,_= image_or.shape
noise = np.random.normal(0,15,(rows,cols,3)).astype(np.uint8)
# Add the noise to the image
noisy_image = image_or + noise
plt.subplot(1, 2, 1)
plt.imshow(image_or)

plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(noisy_image)
plt.title("Noisy Image")
```

```python
kernel = np.ones((6,6))/36
image_filtered = cv2.filter2D(src=noisy_image, ddepth=-1, kernel=kernel)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(image_filtered)
plt.title("Filtered Image")
plt.subplot(1, 2, 2)
plt.imshow(noisy_image)
plt.title("Noisy Image")
#A smaller kernel keeps the image sharp, but filters less noise
```

```python
#Histogram counts number of occurence of a pixel(Darker portions represents lower intensity)
#Intensity Transformation changes an image one pixel at a time
#Histogram Equalization is an algorithm that uses the image's histogram to adjust contrast (.equalizeHist())
#Threshold function applies a threshold to every pixel,it can be used to extract objects from an image(segmentation)
#It converts the image to black and wight ,we can use THRESH_OTSU method to select the value automatically
```

```python
def plot_image(image_1, image_2,title_1="Orignal", title_2="New Image"):
    plt.figure(figsize=(10,10))
    plt.subplot(1, 2, 1)
    plt.imshow(image_1,cmap="gray")
    plt.title(title_1)
    plt.subplot(1, 2, 2)
    plt.imshow(image_2,cmap="gray")
    plt.title(title_2)
    plt.show()
```

```python
hist = cv2.calcHist([img],[0], None, [256], [0,256])
intensity_values = np.array([x for x in range(hist.shape[0])])
plt.bar(intensity_values, hist[:,0], width = 5)
plt.title("Bar histogram")
plt.show()
```

```python
color = ('blue','green','red')
for i,col in enumerate(color):
    histr = cv2.calcHist([image],[i],None,[256],[0,256])
    plt.plot(intensity_values,histr,color = col,label=col+" channel")
    
    plt.xlim([0,256])
plt.legend()
plt.title("Histogram Channels")
plt.show()
```

```python
alpha = 2# Simple contrast control
beta = 0 # Simple brightness control 
new_image = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
```

```python
plot_image(img, new_image, title_1 = "Orignal", title_2 = "brightness control")
```

```python
new_image = cv2.convertScaleAbs(img, alpha=3, beta=-200)
plot_image(img, new_image, "Orignal", "brightness & contrast control")
```
